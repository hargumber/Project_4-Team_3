```python
import pandas as pd
import os

# Root folder containing all subdirectories with CSV files
root_folder_path = "C:/Users/dson.AzureAD/PROJECT_4/Prepared Data/Makeup_by_state"

# Find all CSV files in the root folder and subdirectories
csv_files = []
for dirpath, dirnames, filenames in os.walk(root_folder_path):
    for file in filenames:
        if file.endswith(".csv"):
            csv_files.append(os.path.join(dirpath, file))

# Initialize an empty list to store DataFrames
dataframes = []

# Load each CSV file and append to the DataFrame list
for file_path in csv_files:
    try:
        df = pd.read_csv(file_path)
        dataframes.append(df)
        print(f"Loaded: {file_path}")
    except Exception as e:
        print(f"Failed to load {file_path}: {e}")

# Combine all DataFrames into one
if dataframes:
    combined_df = pd.concat(dataframes, ignore_index=True)

    # Save the combined DataFrame to a new CSV
    output_file = os.path.join(root_folder_path, "Combined_NRI_Table_Counties.csv")
    combined_df.to_csv(output_file, index=False)
    print(f"All files combined and saved to {output_file}")
else:
    print("No files were loaded. Check the folder paths and file names.")
```
